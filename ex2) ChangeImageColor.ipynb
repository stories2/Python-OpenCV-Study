# OpenCV 기본 예제

로드한 이미지의 색깔을 바꾸는 예제 실습

```python
import numpy
import cv2
%matplotlib inline
from matplotlib import pyplot as plt
```

python 에서 필요한 opencv 모듈을 임포트

```python
baseImage = cv2.imread("test.jpg", cv2.IMREAD_COLOR)
```

테스트할 이미지를 로드

```python
rdyToChangeImage = cv2.cvtColor(baseImage, cv2.cv.CV_BGR2RGB)
```

blue 와 red가 바뀌어서 나오므로 스와핑

```python
plt.imshow(rdyToChangeImage)
```

로드한 이미지를 테스트용으로 출력

```python
grayImage = cv2.cvtColor(rdyToChangeImage, cv2.COLOR_RGB2GRAY)
```

흑백 이미지로 변환

```python
plt.imshow(grayImage, cmap="gray")
```
