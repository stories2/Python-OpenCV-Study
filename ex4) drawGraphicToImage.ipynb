# OpenCV 기본 예제

이미지에 그림을 그리는 예제 실습

```python
import numpy
import cv2
%matplotlib inline
from matplotlib import pyplot as plt
```

필요한 모듈 임포트

```python
baseImage = cv2.imread('test.jpg', cv2.COLOR_BGR2RGB)
```

타겟 이미지를 로드할 때 칼라 픽스를 하라고는 했는데 먹히지가 않는다

```python
baseImage = cv2.cvtColor(baseImage, cv2.COLOR_BGR2RGB)
```

그리하여 다시 써줌 찡긋-!

```python
plt.imshow(baseImage)
```

불러온거 이미지 출력

```python
baseImage.shape
```

이미지 정보를 보자

```python
baseImageWidth, baseImageHeight = baseImage.shape[:2]
print baseImageWidth, baseImageHeight
```

그 중 이미지의 크기만 가져옴

```python
halfOfLength = 2
drawingCircleRadius = 100
drawingPositionX = baseImageWidth / halfOfLength
drawingPositionY = baseImageHeight / halfOfLength
colorCodeOfRed = 128
colorCodeOfGreen = 128
colorCodeOfBlue = 128
```

드로잉 하기 전 소스에 숫자를 최소화 하기 위해 변수들 생성

```python
print drawingPositionX, drawingPositionY
```

변수값 테스트

```python
cv2.circle(baseImage, (drawingPositionX, drawingPositionY), drawingCircleRadius, colorCodeOfBlue, colorCodeOfGreen, colorCodeOfRed, 1)
```

타겟 이미지의 특정 좌표에 원을 원하는 색깔로 그려준다

```python
plt.imshow(baseImage)
```
