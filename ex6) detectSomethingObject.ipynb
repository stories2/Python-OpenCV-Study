# OpenCV 기본 예제

영상의 특정 부분을 디텍트 하는 예제를 실습

```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

필요한 모듈 임포트

```python
detectTargetImage = cv2.imread('detectTargetImage.jpg', 0)
```

디텍트할 목표 이미지를 로드

```python
plt.imshow(detectTargetImage, cmap="gray")
```

디텍트할 이미지 로드 후 출력 테스트

```python
detectTargetImage.shape
```

이미지 크기 정보 출력

```python
detectTemplateImage = cv2.imread('detectTemplateImage.jpg', 0)
```

디텍트할 목표 이미지를 찾을 배경 이미지 로드

```python
plt.imshow(detectTemplateImage, cmap="gray")
```

배경 이미지 출력 테스트

```python
detectTemplateImage.shape
```

배경 이미지 크기 정보 출력

```python
detectTargetImageBackup = detectTargetImage.copy()
detectTargetImageBackup.shape
```

디텍트할 목표 이미지 백업

```python
resultOfMethodeTM_CCOEFF = cv2.matchTemplate(detectTargetImage, detectTemplateImage, cv2.TM_CCOEFF)
```

디텍트 하는 여러 메소드 중 TM_CCOEFF라는 메소드를 사용하였는데 이 부분에 대해선 공부가 필요

```python
minVal, maxVal, minLoc, maxLoc = cv2.minMaxLoc(resultOfMethodeTM_CCOEFF)
print minVal, maxVal, minLoc, maxLoc
```

결과값 위치 정보 출력

```python
imageWidthSavedPoint = 0
imageHeightSavedPoint = 1
```

코드 이뻐져라 이뻐져라 얍!

```python
topLeft = maxLoc
bottomRight = (topLeft[imageWidthSavedPoint] + detectTargetImageBackup.shape[imageHeightSavedPoint], 
               topLeft[imageHeightSavedPoint] + detectTargetImageBackup.shape[imageWidthSavedPoint])
print topLeft, bottomRight
```

디텍트 된 목표 이미지가 배경 이미지의 어느 위치에 있는지 알려주기 위해 네모를 그리는 함수에 집어 넣을 위치 정보들의 값들을 계산

```python
cv2.rectangle(detectTemplateImage, topLeft, bottomRight, 0, 1)
```

배경 이미지에 네모를 그리자!

```python
cv2.imwrite("detectedResultImage.jpg", detectTemplateImage)
```

그린 다음 저장

```python
detectTemplateImage = cv2.imread("detectedResultImage.jpg", 0)
```

저장한 결과 이미지를 다시 로드

```python
plt.imshow(detectTemplateImage, cmap="gray")
```
