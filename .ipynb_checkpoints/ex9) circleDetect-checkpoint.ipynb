# OpenCV 기본 예제

영상에서 원을 디텍트하는 예제

```python
%matplotlib inline
import cv2
import numpy as np
from matplotlib import pyplot as plt
import cv2.cv as cv
```

필요한 모듈 로드

```python
targetImage = cv2.imread('resizedTest.jpg', 0)
```

디텍트할 영상을 로드하되 흑백으로

```python
plt.imshow(targetImage, cmap = "gray")
```

로드한 영상 출력

```python
targetImage = cv2.medianBlur(targetImage, 3)
```

영상 블러효과를 적용하여 뭉갠다

```python
plt.imshow(targetImage, cmap = "gray")
```

해당 영상 출력

```python
circleDetection = cv2.HoughCircles(targetImage, cv.CV_HOUGH_GRADIENT, 1, 20, 
                                   param1 = 50, param2 = 30, minRadius = 0, maxRadius = 0)
print circleDetection
```

원모양 추출

```python
circleDetection = np.uint16(np.around(circleDetection))
print circleDetection
```

소수점 버리고 unsigned int16 형으로 변환

```python
for eachCircleData in circleDetection[0, :]:
    cv2.circle(targetImage, (eachCircleData[0], eachCircleData[1]), eachCircleData[2], 0, 2)
```

구해진 좌표와 크기 데이터를 이용해 타겟 영상에 원을 그린다

```python
plt.imshow(targetImage, cmap = "gray")
```

그리고 출력 했는데 좀 많다
