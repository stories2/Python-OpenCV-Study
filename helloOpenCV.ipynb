```python
import cv2
%matplotlib inline
from matplotlib import pyplot as plt
```

```python
print "OpenCV version : %s" % cv2.__version__
```

설치된 opencv 버전 확인

```python
opencvTestImage = cv2.imread("test.jpg")
```

테스트할 이미지를 로드

```python
opencvTestImage = cv2.cvtColor(opencvTestImage, cv2.cv.CV_BGR2RGB)
```

그냥 출력하면 red와 blue가 바뀌어서 나오므로 로드할때 스와핑 하여 문제를 해결
