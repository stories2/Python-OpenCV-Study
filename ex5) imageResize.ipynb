# OpenCV 기본 예제

이미지의 크기를 조절하는 예제

```python
import numpy
import cv2
%matplotlib inline
from matplotlib import pyplot as plt
```

필요한 모듈을 임포트

```python
expectImageWidthSize = 300.0
imageWidthSavedPoint = 0
imageHeightSavedPoint = 1
```

사용할 변수들을 정의

```python
baseImage = cv2.imread('test.jpg')
```

사용할 이미지를 로드

```python
baseImage = cv2.cvtColor(baseImage, cv2.COLOR_BGR2RGB)
```

파란색과 빨간색이 바뀌어서 출력되니 수정

```python
resizedHeight = int((expectImageWidthSize / baseImage.shape[imageWidthSavedPoint]) * baseImage.shape[imageHeightSavedPoint])
resizedWidth = int(expectImageWidthSize)
print resizedWidth, resizedHeight
```

사이즈를 어느정도로 조절할지 비율에 따라서 계산함

```python
resizedImage = cv2.resize(baseImage, (resizedHeight, resizedWidth), interpolation = cv2.INTER_AREA)
```

이미지 사이즈 조절

```python
plt.imshow(resizedImage)
```

조절 후 출력

```python
resizedImage = cv2.cvtColor(resizedImage, cv2.COLOR_RGB2BGR)
```

저장을 할때 다시 바꾸어 놓지 않으면 또 색깔이 뒤바껴서 저장됨

```python
cv2.imwrite('resizedTest.jpg', resizedImage)
```
